```
## checking Null Values for the specific features.
print("Missing Values in 'Region' Column          :",feat["Region"].isnull().sum())
print("Missing Values in 'Children' Column        :",feat["Children"].isnull().sum())
print("Missing Values in 'cases/1M' Column        :",feat["cases/1M"].isnull().sum())
print("Missing Values in 'Married' Column         :",feat["Married"].isnull().sum())
print("Missing Values in 'Age' Column             :",feat["Age"].isnull().sum())
print("Missing Values in 'Coma score' Column      :",feat["Coma score"].isnull().sum())
print("Missing Values in 'Diuresis' Column        :",feat["Diuresis"].isnull().sum())
print("Missing Values in 'Platelets' Column       :",feat["Platelets"].isnull().sum())
print("Missing Values in 'd-dimer' Column         :",feat["d-dimer"].isnull().sum())
print("Missing Values in 'Heart rate' Column      :",feat["Heart rate"].isnull().sum())
print("Missing Values in 'HDL cholesterol' Column :",feat["HDL cholesterol"].isnull().sum())
print("Missing Values in 'Insurance' Column       :",feat["Insurance"].isnull().sum())
print("Missing Values in 'salary' Column          :",feat["salary"].isnull().sum())
print("Missing Values in 'FT/month' Column        :",feat["FT/month"].isnull().sum())

```

```
# Handling Null Values Or Missing Values:
med = feat["Children"].median()
feat["Children"].fillna(med,inplace = True)
print("Missing Values in 'Children' Column after filling with median       :",feat["Children"].isnull().sum())
```

```
med = feat["Diuresis"].mean()
feat["Diuresis"].fillna(med,inplace = True)
print("Missing Values in 'Diuresis' Column after filling with mean       :",feat["Diuresis"].isnull().sum())
```

```
med = feat["Platelets"].median()
feat["Platelets"].fillna(med,inplace = True)
print("Missing Values in 'Platelets' Column after filling with median       :",feat["Platelets"].isnull().sum())
```

```
med = feat["d-dimer"].median()
feat["d-dimer"].fillna(med,inplace = True)
print("Missing Values in 'd-dimer' Column after filling with median       :",feat["d-dimer"].isnull().sum())
```

```
med = feat["Heart rate"].mean()
feat["Heart rate"].fillna(med,inplace = True)
print("Missing Values in 'Heart rate' Column after filling with mean       :",feat["Heart rate"].isnull().sum())
```

```
med = feat["HDL cholesterol"].median()
feat["HDL cholesterol"].fillna(med,inplace = True)
print("Missing Values in 'HDL cholesterol' Column after filling with median       :",feat["HDL cholesterol"].isnull().sum())
```

```
med = feat["Insurance"].mean()
feat["Insurance"].fillna(med,inplace = True)
print("Missing Values in 'Insurance' Column after filling with median       :",feat["Insurance"].isnull().sum())
```

```
med = feat["FT/month"].median()
feat["FT/month"].fillna(med,inplace = True)
print("Missing Values in 'FT/month' Column after filling with median       :",feat["FT/month"].isnull().sum())
```

```
# LabelEncoing for Catogorical variable
 from sklearn.preprocessing import LabelEncoder
 from sklearn.preprocessing import OneHotEncoder

 encoder = LabelEncoder()
 onehot  = OneHotEncoder(sparse=False)

 feat["Married"] = encoder.fit_transform(feat["Married"])
 feat["Region"] = encoder.fit_transform(feat["Region"])
```

```
##finding Outliers in the data

from xgboost import XGBRegressor
```

```
my_model = XGBRegressor()
```

```
features = feat.loc[:,final_cols]
tar = pd.DataFrame(tar)
```

```
# train and test Spliting
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(features,tar,test_size = 0.3)
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
my_model.fit(x_train,y_train,verbose=False)
```

```
predictions_2 = my_model.predict(x_test)
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error for n_estimators  = 100 : " + str(mean_absolute_error(predictions_2, y_test)))
```

```
#model tuning 

# making n_estimators  =  1000

my_model2 = XGBRegressor(n_estimators = 73)

my_model2.fit(x_train,y_train,verbose = False),
predictions_3 = my_model2.predict(x_test)
from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error for n_estimators  = 73 : " + str(mean_absolute_error(predictions_3, y_test)))
```
